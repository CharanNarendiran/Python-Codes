```python
def time_mark(time):
    if int(time[13:15]) < 9:
        return 1
    elif int(time[13:15]) < 11:
        return 2
    elif int(time[13:15]) < 13:
        return 3
    return 4
```

```python
sample['punc_mark'] = sample['Timestamp'].apply(lambda time : time_mark(time))
```

```python
def prog_level(num):
    if num == 1:
        return 'Low'
    elif 1 < num < 4:
        return 'Medium'
    return 'Good'

sample['Programming level'] = sample['Programming level'].apply(lambda num : prog_level(num))
sample.head(15)
```
